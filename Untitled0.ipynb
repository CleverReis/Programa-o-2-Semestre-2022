```python
import matplotlib.pyplot as plt

x1 = [1, 3, 5, 7, 9]
y1 = [2, 6, 10, 14, 18]

x2 = [2, 4, 6, 8, 10]
y2 = [4, 8, 12, 16, 20]

titulo = "Primeiro Grafico com Python"
eixox = "Eixo X"
eixoy = "Eixo Y"

plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.bar(x1,y1, label = "Grupo1", color = "black")
plt.bar(x2,y2, label = "Grupo2", color = 'orange')
plt.legend()
plt.show()




```
